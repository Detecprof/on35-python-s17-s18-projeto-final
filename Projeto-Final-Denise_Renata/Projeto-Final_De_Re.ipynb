```python
dfSaresp.columns = dfSaresp.columns.str.strip()
```

```python
dfSaresp.columns = ['DEPADM', 
', DEPBOL', 'NOMEDEPBOL', 'CODRMET', 'CODMUN', 'MUN', 'SERIE_ANO', 'COD_PER', 'PERIODO', 'CO_COMP', 'DS_COMP', 'MEDPROF']
```

```python
dfSaresp.columns
```

```python
dfSaresp.head()
```

```python

dfSaresp = dfSaresp.drop(['DEPADM', ', DEPBOL', 'CODRMET', 'CODMUN', 'COD_PER', 'CO_COMP'], axis=1)
```

```python
dfSaresp.head()
```

```python
dfSaresp = dfSaresp.rename(columns={
    'NOMEDEPBOL': 'REDE DE ENSINO', 
    'MUN': 'MUNICIPIO', 
    'DS_COMP': 'COMPONENTE_CURRICULAR', 
    'MEDPROF': 'MEDIA_PROFICIENCIA'
})
```

```python
dfSaresp.head()
```

```python
dfSaresp.dtypes
```

```python
# Substituir vírgulas por pontos para interpretar como decimais
dfSaresp['MEDIA_PROFICIENCIA'] = dfSaresp['MEDIA_PROFICIENCIA'].str.replace(',', '.')

# Converter a coluna para numérico (float)
dfSaresp['MEDIA_PROFICIENCIA'] = pd.to_numeric(dfSaresp['MEDIA_PROFICIENCIA'], errors='coerce')
```

```python
# Valor mínimo considerando apenas valores maiores que 100
minimo = dfSaresp['MEDIA_PROFICIENCIA'][dfSaresp['MEDIA_PROFICIENCIA'] > 100].min()

# Valor máximo
maximo = dfSaresp['MEDIA_PROFICIENCIA'].max()

# Soma dos valores
soma = dfSaresp['MEDIA_PROFICIENCIA'].sum()

# Média dos valores
media = dfSaresp['MEDIA_PROFICIENCIA'].mean()

print(f"""Menor Média Saresp 2023: {minimo}
Maior Média Saresp 2023: {maximo}
Soma das Médias: {soma}
Média de Proficiência Saresp: {media}""")
```

```python
import matplotlib.pyplot as plt

# Dados a serem plotados (média, máximo, mínimo)
valores = [media, maximo, minimo]
labels = ['Média', 'Máximo', 'Mínimo']

# Criar o gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(labels, valores, color=['skyblue', 'green', 'red'])

# Adicionar título e rótulos
plt.title('Média, Máximo e Mínimo da Proficiência Saresp 2023')
plt.ylabel('Valores')

# Exibir o gráfico
plt.show()
```

```python
dfCeara.head()
```

```python
# Exibe os nomes das colunas do DataFrame para verificação
print("Colunas existentes antes da remoção:")
print(dfCeara.columns)

# Lista das colunas que você gostaria de excluir
colunas_para_remover = [
    'Código da Etapa', 
    'Código da Rede', 
    'Código do Município', 
    'Taxa de Participação', 
    'Proficiência Padronizada', 
    'Fator de Ajuste', 
    'Número de Alunos no Muito', 
    'Número de Alunos no Crítico', 
    'Número de Alunos no Intermediário', 
    'Número de Alunos no Adequado', 
    'IDE EF MT', 
    'IDE Médio EF'
]

# Filtra as colunas que realmente existem no DataFrame
colunas_existentes = [col for col in colunas_para_remover if col in dfCeara.columns]

# Exibe as colunas que serão removidas
print("Colunas que serão removidas:")
print(colunas_existentes)

# Remove as colunas existentes
dfCeara = dfCeara.drop(colunas_existentes, axis=1)

# Exibe os nomes das colunas após a remoção para verificação
print("Colunas existentes após a remoção:")
print(dfCeara.columns)
```

```python
dfCeara.head()
```

```python
# Renomeia as colunas no DataFrame dfCeara
dfCeara = dfCeara.rename(columns={
    'Etapa': 'SERIE_ANO', 
    'Rede': 'REDE DE ENSINO', 
    'Município': 'MUNICIPIO', 
    'Edição': 'ANO_PROVA', 
    'Proficiência Média': 'MEDIA_PROFICIENCIA'
})

# Exibe os nomes das colunas após o renomeio para verificação
print("Colunas após o renomeio:")
print(dfCeara.columns)
```

```python
dfCeara.head()
```

```python
# Substituir vírgulas por pontos para interpretar como decimais
dfCeara['MEDIA_PROFICIENCIA'] = dfCeara['MEDIA_PROFICIENCIA'].str.replace(',', '.')

# Converter a coluna para numérico (float)
dfCeara['MEDIA_PROFICIENCIA'] = pd.to_numeric(dfCeara['MEDIA_PROFICIENCIA'], errors='coerce')
```

```python
# Valor mínimo considerando apenas valores maiores que 100
minimo = dfCeara['MEDIA_PROFICIENCIA'][dfCeara['MEDIA_PROFICIENCIA'] > 100].min()

# Valor máximo
maximo = dfCeara['MEDIA_PROFICIENCIA'].max()

# Soma dos valores
soma = dfCeara['MEDIA_PROFICIENCIA'].sum()

# Média dos valores
media = dfCeara['MEDIA_PROFICIENCIA'].mean()

print(f"""Menor Média Ceará: {minimo}
Maior Média Ceará: {maximo}
Soma das Médias: {soma}
Média de Proficiência Ceará 2023: {media}""")
```
