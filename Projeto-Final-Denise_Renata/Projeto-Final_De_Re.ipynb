```python
import pandas as pd
dfSaresp = pd.read_csv('Dados_Saresp_2023.csv',delimiter=';', encoding='utf-8', skiprows=8)
```

```python
dfSaresp.head()
```

```python
dfSaresp.tail()
```

```python
print(dfSaresp.shape)
```

```python
nulos_por_coluna = dfSaresp.isnull().sum()
print(nulos_por_coluna)
```

```python
dfSaresp.columns
```

```python
dfSaresp.columns = dfSaresp.columns.str.strip()
```

```python
dfSaresp.columns = ['DEPADM', 
', DEPBOL', 'NOMEDEPBOL', 'CODRMET', 'CODMUN', 'MUN', 'SERIE_ANO', 'COD_PER', 'PERIODO', 'CO_COMP', 'DS_COMP', 'MEDPROF']
```

```python
dfSaresp.columns
```

```python
dfSaresp.head()
```

```python

dfSaresp = dfSaresp.drop(['DEPADM', ', DEPBOL', 'CODRMET', 'CODMUN', 'COD_PER', 'CO_COMP'], axis=1)
```

```python
dfSaresp.head()
```

```python
dfSaresp = dfSaresp.rename(columns={
    'NOMEDEPBOL': 'REDE DE ENSINO', 
    'MUN': 'MUNICIPIO', 
    'DS_COMP': 'COMPONENTE_CURRICULAR', 
    'CODMUN': 'CODIGO MUNICIPIO',
    'SERIE_ANO': 'SERIE_ANO',
    'PERIODO': 'PERIODO', 
    'MEDPROF': 'MEDIA_PROFICIENCIA'
})
```

```python
dfSaresp.head()
```

```python
dfSaresp.dtypes
```

```python
# Substituir vírgulas por pontos para interpretar como decimais
dfSaresp['MEDIA_PROFICIENCIA'] = dfSaresp['MEDIA_PROFICIENCIA'].str.replace(',', '.')

# Converter a coluna para numérico (float)
dfSaresp['MEDIA_PROFICIENCIA'] = pd.to_numeric(dfSaresp['MEDIA_PROFICIENCIA'], errors='coerce')
```

```python
# Valor mínimo considerando apenas valores maiores que 100
minimo = dfSaresp['MEDIA_PROFICIENCIA'][dfSaresp['MEDIA_PROFICIENCIA'] > 100].min()

# Valor máximo
maximo = dfSaresp['MEDIA_PROFICIENCIA'].max()

# Soma dos valores
soma = dfSaresp['MEDIA_PROFICIENCIA'].sum()

# Média dos valores
media = dfSaresp['MEDIA_PROFICIENCIA'].mean()

print(f"""Menor Média Saresp 2023: {minimo}
Maior Média Saresp 2023: {maximo}
Soma das Médias: {soma}
Média de Proficiência Saresp: {media}""")
```

```python
import matplotlib.pyplot as plt

# Dados a serem plotados (média, máximo, mínimo)
valores = [media, maximo, minimo]
labels = ['Média', 'Máximo', 'Mínimo']

# Criar o gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(labels, valores, color=['skyblue', 'green', 'red'])

# Adicionar título e rótulos
plt.title('Média, Máximo e Mínimo da Proficiência Saresp 2023')
plt.ylabel('Valores')

# Exibir o gráfico
plt.show()
```

```python
# Calcular a média de proficiência por tipo de rede
media_por_rede = dfSaresp.groupby('REDE DE ENSINO')['MEDIA_PROFICIENCIA'].mean()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar gráfico de barras
plt.figure(figsize=(8, 5))
media_por_rede.plot(kind='bar', color=['blue', 'orange'])

# Adicionar título e rótulos
plt.title('Comparação da Média de Proficiência SARESP 2023: Rede Estadual vs Rede Municipal')
plt.ylabel('Média de Proficiência por Rede de Ensino')
plt.xlabel('MODADALIDADE DE REDE DE ENSINO')

# Adicionar valores em cima das barras
for index, value in enumerate(media_por_rede):
    plt.text(index, value, round(value, 2), ha='center', va='bottom')

# Exibir o gráfico
plt.xticks(rotation=0)  # Para manter os rótulos na horizontal
plt.tight_layout()  # Para ajustar o layout
plt.show()
```
